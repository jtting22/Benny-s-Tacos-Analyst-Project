**Word Cloud**

Word clouds are especially effective for us because we are looking at review data.  Below you can see the various sentiments with and without stop words from the 90013 zipcode.

```
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt

# concatenate all the reviews into one single string 
full_text = ' '.join(yelp_90013['text'])

my_stop_words = ["nice", "little", "amazing", "restaurant","delicious","love","really", "loved","Grand","first","Central","best","time","great","good","one","Grand-Central","go","got","place", "food","https", "co", "RT", 'aren', 'couldn', 'didn', 'doesn', 'don', 'hadn', 'hasn', 'haven', 'isn', 'let', 'll', 'mustn', 're', 'rt', 'shan', 'shouldn', 've', 'wasn', 'weren', 'won', 'wouldn'] + list(STOPWORDS)
cloud_no_stopword = WordCloud(width=1600, height=800,max_font_size=200, background_color='white', stopwords=my_stop_words).generate(full_text)

plt.imshow(cloud_no_stopword, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```
#Importing packages
from wordcloud import WordCloud
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
```

```
#Build a wordcloud that removes stopwords on a darker background
my_stop_words = ["nice", "little", "amazing", "restaurant","delicious","love","really", "loved","Grand","first","Central","best","time","great","good","one","Grand-Central","go","got","place", "food","https", "co", "RT", 'aren', 'couldn', 'didn', 'doesn', 'don', 'hadn', 'hasn', 'haven', 'isn', 'let', 'll', 'mustn', 're', 'rt', 'shan', 'shouldn', 've', 'wasn', 'weren', 'won', 'wouldn'] + list(STOPWORDS)
def wordcloud(df, posneg):
  pn_string = []
  for t in yelp_90013['text']:
    pn_string.append(t)
  pn_string = pd.Series(pn_string).str.cat(sep=' ')
  if posneg == 1:
    c = 'autumn'
    bc = 'black'
  else:
    c = 'summer'
    bc = 'white'
  wordcloud = WordCloud(width=1600, height=800,max_font_size=200,colormap=c,background_color=bc,stopwords=my_stop_words).generate(pn_string) 
  plt.figure(figsize=(12,10))
  plt.imshow(wordcloud, interpolation="bilinear")
  plt.axis("off")
  plt.show()
```

```
wordcloud(yelp_90013,1)
```
